```python
df = df.loc[0:8523,:].dropna(how = 'any')
```

```python
dairy = df['Item_Type'] == 'Dairy'
dairy
```

```python
lowFat = df['Item_Fat_Content'] == 'Low Fat'
```

```python
lowFat_dairy = df.loc[dairy & lowFat]
lowFat_dairy
```

```python
regular_fat = df['Item_Fat_Content'] == 'Regular'
```

```python
regular_dairy = df.loc[dairy & regular_fat]
regular_dairy
```

```python
lowFat_dairy['Item_MRP'].cumsum()
```

```python
lowFat_dairy['Item_Outlet_Sales'].cumsum()
```

```python
regular_dairy['Item_MRP'].cumsum()
```

```python
regular_dairy['Item_Outlet_Sales'].cumsum()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (10, 5))
axes.plot(lowFat_dairy['Item_Outlet_Sales'].cumsum(), lowFat_dairy['Item_MRP'].cumsum(), color = 'red', label = 'Low Fat Dairy MRP');
axes.plot(regular_dairy['Item_Outlet_Sales'].cumsum(), regular_dairy['Item_MRP'].cumsum(), color = 'blue', label = 'Rgular Dairy MRP');
axes.set_xlim(left=0,right=300000);
axes.set_ylim(bottom=0,top=25000);
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Outlet Sales', fontsize = 15);
plt.ylabel('Item MRP', fontsize = 15);
plt.title('Outlet sales per Dairy MRP', fontsize = 15)
axes.legend(loc="lower right", fontsize = 15);
```

```python
df['Item_MRP'].hist(bins = 30,
                                   edgecolor='black', color='cyan');
plt.ticklabel_format(useOffset=False, style='plain')
plt.xlabel('Item MRP', fontsize = 15);
plt.ylabel('Items', fontsize = 15);
plt.title('MRP by Number of Items', fontsize = 15);
```

```python
fat_dummies = pd.get_dummies(df['Item_Fat_Content'])
fat_dummies.head()
```

```python
df = pd.concat([df, fat_dummies] , axis = 1)
df.head()
```
